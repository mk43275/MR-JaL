```python
import pandas as pd

df = pd.read_csv('/content/text.csv')
df['Sentence'] = df['Sentence'].str.strip()

df['file'] = df['ID'] + '.wav'
```

```python
def createFilePath(fileName):
  path = '/content/\temp/speech_wav'
  p = fileName.split('_')[0]
  final_path = os.path.join(path, p, p + '_' + fileName.split('_')[1])

  return final_path
```

```python
df['file'] = df['file'].apply(createFilePath)

df
```

```python
ls = df['file'].to_list()
trans = []

for sent in ls:
  try:
    text = transcriber(sent)
    text = list(text.values())[0]
    trans.append(text)
  except:
    trans.append('broke')

true_sentences = df['Sentence'].to_list()
```

```python
import nltk
# Download the 'punkt_tab' resource
nltk.download('punkt_tab')
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.tokenize import word_tokenize

# Tokenize the sentences
references = [word_tokenize(s) for s in true_sentences]
candidates = [word_tokenize(s) for s in trans]

# Initialize a smoothing function to handle cases with zero n-gram overlap
smoothing_function = SmoothingFunction().method1

# Calculate BLEU score for each pair of sentences
bleu_scores = []
for i in range(len(true_sentences)):
    reference = [references[i]]  # BLEU expects a list of reference sentences
    candidate = candidates[i]
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=smoothing_function)
    bleu_scores.append(bleu_score)

# Calculate the overall BLEU score (e.g., the average)
overall_bleu = sum(bleu_scores) / len(bleu_scores) if bleu_scores else 0
print(f"\nOverall BLEU score: {overall_bleu:.4f}")
```
